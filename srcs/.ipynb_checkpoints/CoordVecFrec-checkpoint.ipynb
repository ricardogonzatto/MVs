```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
```

```python
# O retorno da função hp.pix2ang() é theta, phi
# Ambas as listas possuem valores repetidos pq mantemos hora mantemos theta fixo para variar phi
# hora variamos phi mantendo theta fixo

nside = 8
print(hp.nside2npix(nside))

coord = []
for i in range (hp.nside2npix(nside)):
    m = hp.pix2ang(8, i)
    coord.append(m)

n = np.transpose(coord)     
#n[0]
np.savetxt("../datas/test1.dat", np.column_stack([n[0], n[1]]))
print(n[1])
```
